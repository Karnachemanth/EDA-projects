```python
#Research question
#Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?)
```

```python
# for that we need to seperate male and female from the gender 
```

```python
df[df['Gender']=='m']['Specialization'].value_counts()
```

```python
df[df['Gender']=='f']['Specialization'].value_counts()
```

```python
#Its unfair to compare male & female based on number of students as the dataset contains 75 percent male and remaining 25 percent female data
# As we can see from the above piechart maybe from the percentages we can come to some conclusions
```

```python
counts_1=df[df['Gender']=='m']['Specialization'].value_counts()
percent_1= counts_1/counts_1.sum()*100
percent_1
```

```python
counts_2=df[df['Gender']=='f']['Specialization'].value_counts()
percent_2= counts_2/counts_2.sum()*100
percent_2
```

```python
df_m = {
    'Specialization': percent_1.index,
    'Percent_m': percent_1.values}
```

```python
df_m=pd.DataFrame(df_m)
```

```python
df_f = {
    'Specialization': percent_2.index,
    'Percent_f': percent_2.values}
```

```python
df_f=pd.DataFrame(df_f)
```

```python
df_new=pd.merge(df_m,df_f,on='Specialization',how='inner')
```

```python
df_new.set_index("Specialization",inplace=True)
```

```python
df_new.plot(kind='bar',figsize=(8,5))
plt.ylabel('Percentage')
plt.show()
```

```python
#As what we see from the plot we can say that women are slightly inclined towards preferring computer sciences related streams greater than men
#while certain streams like mechanical and other core streams men are dominant there compared to women
#Overall we can say that both men & women are focussed on doing computer related graduation irrespective of gender.
```
